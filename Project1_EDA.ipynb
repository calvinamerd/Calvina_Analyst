```python
# Pilih kolom-kolom numerik dari DataFrame
numeric_cols = df.select_dtypes(include=[int, float]).columns

# Lakukan groupby dan hitung rata-rata
df_grouped = df.groupby('Continent')[numeric_cols].mean()

# Sortir berdasarkan kolom '2022 Population'
df_grouped_sorted = df_grouped.sort_values(by="2022 Population", ascending=False)

print(df_grouped_sorted)
```
