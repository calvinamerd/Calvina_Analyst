```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_albums_in_South_Korea#Best-selling_album_by_year'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[3]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[3]
```

```python
print(table)
```

```python
selected_titles = table.find_all('th')
```

```python
selected_titles
```

```python
selected_table_titles = [title.text.strip() for title in selected_titles]

print(selected_table_titles)
```
